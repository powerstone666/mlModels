```python
from ibm_watson import NaturalLanguageUnderstandingV1
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator
from ibm_watson.natural_language_understanding_v1 import Features, EntitiesOptions, KeywordsOptions, CategoriesOptions
from google.colab import userdata

authenticator=IAMAuthenticator(userdata.get('ibm_api'))
nlp = NaturalLanguageUnderstandingV1(
    version='2023-03-01',
    authenticator=authenticator
)
nlp.set_service_url(userdata.get('ibm_url'))
```

```python
new_data.head()
```

```python
def get_watson_features(text):
    try:
        response = nlp.analyze(
            text=text,
            features=Features(
                keywords=KeywordsOptions(limit=3),  # Extract top 3 keywords
                entities=EntitiesOptions(limit=3),  # Extract top 3 entities
                categories=CategoriesOptions()      # Extract categories
            )
        ).get_result()

        # Extract key features from Watson's response
        keywords = [kw['text'] for kw in response.get('keywords', [])]
        entities = [ent['text'] for ent in response.get('entities', [])]
        categories = [cat['label'] for cat in response.get('categories', [])]
        return keywords, entities, categories

    except Exception as e:
        print(f"Error processing text: {text}, Error: {e}")
        return [], [], []
def get_context(text):
    return f"This Product Name is: {text}"

n = 100
watson_results = []  # List to store results for each row
for index, row in new_data.iloc[:1000].iterrows():
    keywords, entities, categories = get_watson_features(get_context(row['Product_Name']))
    watson_results.append({
        'Product_Name': row['Product_Name'],
        'Keywords': keywords,
        'Entities': entities,
        'Categories': categories
    })

watson_data = pd.DataFrame(watson_results)
watson_data.tail()
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
def join_features(row):
    keyword_str = " ".join(row["Keywords"]) if row["Keywords"] else ""
    entity_str = " ".join(row["Entities"]) if row["Entities"] else ""
    category_str = " ".join(row["Categories"]) if row["Categories"] else ""
    return f"{keyword_str} {entity_str} {category_str}".strip()  # Remove leading/trailing spaces

watson_data["Feature_Text"] = watson_data.apply(join_features, axis=1)

# Step 2: Convert Text Features into TF-IDF Vectors
tfidf_vectorizer = TfidfVectorizer()  # Initialize TF-IDF Vectorizer
tfidf_matrix = tfidf_vectorizer.fit_transform(watson_data["Feature_Text"])  # Transform text to TF-IDF matrix

# Step 3: Calculate Cosine Similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)  # Pairwise similarity on all products

# Step 4: Recommendation Function
def recommend_products(product_name, similarity_matrix, data_frame, top_n=3):

    product_index = data_frame[data_frame["Product_Name"] == product_name].index[0]

    similarity_scores = list(enumerate(similarity_matrix[product_index]))

    # Sort products by similarity score (excluding itself)
    sorted_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)[1:]

    # Get the indices of the most similar products
    recommended_indices = [x[0] for x in sorted_scores[:top_n]]

    # Return the recommended products
    return data_frame.iloc[recommended_indices]

```

```python
import gradio as gr


def get_Items(product_name):
  target_product = product_name.strip()
  recommended_products = recommend_products(target_product, cosine_sim, watson_data, top_n=5)

# Display Recommended Products
  print(f"Recommendations for '{target_product}':")
  recommendation_list=[]
  for row in recommended_products.itertuples():
     product_info = new_data.loc[new_data['Product_Name'] == row.Product_Name]
     if not product_info.empty:
      recommendation_list.append(product_info.iloc[0].to_dict())
  product=pd.DataFrame(recommendation_list)
  return product



# Gradio UI
iface = gr.Interface(
    fn=get_Items,
    inputs=gr.Textbox(label="Enter Product Name"),
    outputs=gr.DataFrame(headers=["Product_Name","Brand","Package_price","Category","Sub_Category","Product_Group"],type="pandas"),
    title="Product Recommendation System",
    description="Enter a product name to get relevant product recommendations.",
)

# Launch Gradio App
iface.launch()
```
