```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("jackksoncsie/spam-email-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("/root/.cache/kagglehub/datasets/jackksoncsie/spam-email-dataset/versions/1/emails.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
notspam=data[data["spam"]==0]
spam=data[data["spam"]==1]
print(notspam)
```

```python
fig=px.histogram(data,x="spam",color="spam",title="spam email")
fig.update_layout(bargap=0.2)
fig.show()
```

```python
vectorizer=TfidfVectorizer()
features=vectorizer.fit_transform(data["text"])
```

```python
x=features
y=data['spam']
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
new_text=["gift from imran for birthday"]
new_vect=vectorizer.transform(new_text)
prediction=model.predict(new_vect)
if prediction[0]==0:
  print("not spam")
else:
  print("spam")
```

```python
y_pred=model.predict(X_test)
accuracy=accuracy_score(Y_test,y_pred)
print(accuracy)
```
