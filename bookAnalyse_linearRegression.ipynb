```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import plotly.express as px
```

```python
data=pd.read_csv("/content/sleep and psychological effects.csv")
print(data)
```

```python
fig=px.histogram(data,x='Mood_Impact',marginal='box',title="sleep vs mood",color='Weekly_Reading_Time(hours)')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
data.head(10)
```

```python
gender={'f':0,'m':1}
data['Gender']=data['Gender'].map(gender)
data.head(10)
```

```python
data['Mood_Impact']=data['Mood_Impact'].replace('Neutral','Negative')
impact={'Positive':1,'Negative':0}
data['Mood_Impact']=data['Mood_Impact'].map(impact)

data.head(10)
```

```python
book_maping= {book: i + 1 for i, book in enumerate(data['Favorite_Book_Genre'].unique())}
data['Favorite_Book_Genre']=data['Favorite_Book_Genre'].map(book_maping)
data.head(20)
```

```python
model=LinearRegression()
input=data[['Age','Gender','Favorite_Book_Genre','Weekly_Reading_Time(hours)']]
output=data['Mood_Impact']
model.fit(input,output)
```

```python
print(data.head(10))
new_data_point = pd.DataFrame({
    'Age': [32],
    'Gender': [1],
    'Favorite_Book_Genre': [2],
    'Weekly_Reading_Time(hours)': [2]
})
prediction = model.predict(new_data_point)
print(np.round(prediction),"Actual",prediction)
```

```python
accuracy=np.sqrt(mean_squared_error(output,prediction))
print(accuracy)
```
